```python
import itertools
import pandas as pd
df_emnlp = df_2015_target.loc[
    df_2015_target['venue'] == 'Conference on Empirical Methods in Natural Language Processing',
    ['abstract', 'venue']
].drop_duplicates()
df_eccv = df_2015_target.loc[
    df_2015_target['venue'] == 'Computer Vision and Pattern Recognition',
    ['abstract', 'venue']
].drop_duplicates()
pairs = list(itertools.product(df_emnlp.itertuples(index=False), df_eccv.itertuples(index=False)))
df_pairs = pd.DataFrame(
    [(a.abstract, a.venue, b.abstract, b.venue) for a, b in pairs],
    columns=['abstract_1', 'venue_1', 'abstract_2', 'venue_2']
)
df_pairs_unique = df_pairs.apply(
    lambda row: tuple(sorted(
        [(row['abstract_1'], row['venue_1']), (row['abstract_2'], row['venue_2'])],
        key=lambda x: x[0]  
    )),
    axis=1
).drop_duplicates()
df_pairs_unique = pd.DataFrame(
    [(a[0], a[1], b[0], b[1]) for a, b in df_pairs_unique.tolist()],
    columns=['abstract_1', 'venue_1', 'abstract_2', 'venue_2']
)
df_pairs_unique = df_pairs_unique.applymap(lambda x: x.strip().lower() if isinstance(x, str) else x)
df_pairs_unique = df_pairs_unique[
    (df_pairs_unique['abstract_1'].str.split().str.len() > 20) &
    (df_pairs_unique['abstract_2'].str.split().str.len() > 20)
]
df_pairs_unique = df_pairs_unique.sample(1000, random_state=42)
df_pairs_unique.to_csv('nlp_cv.csv',index=False)
```

```python
import pandas as pd
df=pd.read_csv('nlp_cv.csv')
cols_to_clean = ['raw', 'one_shot', 'got', 'cot', 'tot', 'few_shot']
for col in cols_to_clean:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r'^(?i)abstract[:.]?\s*', '', regex=True)  
        .str.replace(r'\n', ' ', regex=True)          
        .str.replace(r'\s+', ' ', regex=True)          
        .str.strip()  
        .str.lower()                                  
    )
df.to_csv('nlp_cv.csv',index=False)
```
